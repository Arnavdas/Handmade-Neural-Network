# To classify only multiples of 10

```python
import numpy as np    
import random
%matplotlib inline
import matplotlib.pyplot as plt

# setting up the neural architecture
def nn_construct():
    print('no of layers '); nn_arch = []
    layers = int(input())
    for i in range(layers):
        print('neurons in the layer ',i+1)
        nn_arch.append({'input_dim':int(input())})
        
        if i:
            nn_arch[i-1]['output_dim'] = nn_arch[i]['input_dim']

    nn_arch[layers - 1]['output_dim'] = nn_arch[layers - 1]['input_dim']

    return nn_arch

# necessary constituent functions for neural network 
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def initialize(layers, nn_arch):
    w, b= {}, {}
    for i in range(1, layers):
        w[i] = np.random.randn(nn_arch[i-1]['output_dim'], nn_arch[i-1]['input_dim'])*0.1
        b[i] = np.random.randn(nn_arch[i-1]['output_dim'], 1)*0.1

    return w, b

def feed_forward(w, a, b, i):
    return np.add(np.dot(w[i], a[i-1]), b[i])

def sigmoid_prime(a):
    return a*(1-a)

def back_ward(y, w, a, dcost, sigmoid_prime, i):
    m = a[i-1].shape[0]

    if len(dcost) == 0:# if it's last layer of the network 
        dcost[i] = a[i] - y 
    else:
        dcost[i] = np.multiply(np.dot(w[i+1].T, dcost[i+1]), sigmoid_prime(a[i]))

    dwt = np.dot(a[i-1], dcost[i].T)
    db = dcost[i]

    return dwt, db, dcost
    
def update(w, b, dw, db, i, n = 1, eta = 0.01):
    if type(i)==int:# for SGD
        w[i] -= (eta*dw)/n
        b[i] = np.subtract(b[i], (eta*db)/n)

    else:# for MGD
        for j in w.keys():
            w[j] -= eta*i[0][j]/n
            b[j] = np.subtract(b[j], (eta*i[1][j])/n)

    return w, b

def model_trainer(gd, epochs, nn_arch, batches = 10, data_size = 200):
    activate = {}; z = {}; samp = []
    dcost = {}; dwt = {}; db = {}; layers = len(nn_arch)

    weights, bias = initialize(layers, nn_arch)# initializing with random real nos in (1,0)
    
    for f in range(epochs):
        
        data = {i : 1 if i%10==0 else 0 for i in random.sample(range(1000, 5000), data_size)}# trainning data preparation

        if gd == 'sgd':# stochastic gradient descent

            for i in data.keys():# for each data i in training data x, we train the layers
                activate[0] = np.array(list(map(int ,str(i)))).reshape(nn_arch[0]['input_dim'], 1)#reshaping the input to 1st layer dim
                
                for j in range(1, layers):# feedforwarding through all layers
                    z[j] = feed_forward(weights, activate, bias, j)# z = wa(of prev layer) + b
                    activate[j] = sigmoid(z[j])# activation = sigma(z)

                for j in range(layers-1, 0, -1):# backpropagating through all the layers
                    dwt, db, dcost = back_ward(np.array([data[i]]), weights, activate, dcost, sigmoid_prime, j)
                    weights, bias = update(weights, bias, dwt.T, db, j)# updating bias and weights
                
                dcost.clear()#clear variables dcost and z for next iteration
                z.clear()

        elif gd == 'mgd':# mini batch gradient descent

            while len(samp) < len(data.keys()):# sample all the data points present in the data
                nw, nb = {k:np.zeros(weights[k].shape) for k in weights.keys()}, {k:np.zeros(bias[k].shape) for k in bias.keys()}
                samp.extend(random.sample(set(data.keys()).difference(samp), batches))# sampling 10 data points

                for i in samp:# for each data i in training data x, we train the layers
                    activate[0] = np.array(list(map(int ,str(i)))).reshape(nn_arch[0]['input_dim'], 1)#reshaping the input to 1st layer dim
                    
                    for j in range(1, layers):# feedforwarding through all layers
                        z[j] = feed_forward(weights, activate, bias, j)# z = wa(of prev layer) + b
                        activate[j] = sigmoid(z[j])# activation = sigma(z)

                    for j in range(layers-1, 0, -1):# backpropagating through all the layers
                        dwt, db, dcost = back_ward(np.array([data[i]]), weights, activate, dcost, sigmoid_prime, j)
                        nw[j] += dwt.T
                        nb[j] += db

                    dcost.clear()#clear variables dcost and z for next iteration
                    z.clear()

                weights, bias = update(weights, bias, dwt.T, db, (nw, nb), batches)
            
    return weights, bias
```

```python
def model_predict(w, b, x_input, L, thresh, input_dim):
    activate = {}; z = {}
    activate[0] = np.array(list(map(int ,str(x_input)))).reshape(input_dim, 1)
    
    for j in range(1, L):
        z[j] = feed_forward(w, activate, b, j)
        activate[j] = sigmoid(z[j])
    
    if thresh == 1:# just to see where the actual activations lighten up for the last layer
        return activate[len(activate) - 1][0] 
    else:
        if activate[len(activate)-1][0] >= thresh:
            return activate[len(activate) - 1][0], 1
        else:
            return activate[len(activate) - 1][0], 0

tru = lambda d : 1 if d%10==0 else 0 # truth value 

def accuracy(real, out, tn, tp, fp, fn):
    pr, rc = None, None
    if tp+fp !=0:
        pr = tp/(tp+fp)
    if tp+fn != 0:
        rc = tp/(tp+fn)
    print('accuracy : ', (tp+fn)/len(real))
    print('recall (True Positive rate) : ',rc)
    print('specificty (True Negative Rate) : ',tn/(tn+fp))
    print('precision (Positive Predictive Value ) : ',pr)
    if pr != None and rc != None:
        print('f1 (balance between precison & recall) : ',2*(1/(1/pr)+(1/rc)))
    
def testing(weights, bias, L, thresh, input_dim=4):
    testing_x  = random.sample(range(5000, 7000), 200)# input your testing data
    mapp = {i: model_predict(weights, bias, i, L, thresh, input_dim) for i in testing_x}
    return mapp
```

```python
def graph(mapp, thresh):
    kk = sorted(mapp.keys())
    if thresh != 1:
        x_cord = lambda g : list(map(lambda t : t, list(filter(lambda s : mapp[s][1] == g == tru(s), kk))))# red for true non-10s
        y_cord = lambda x : list(map(lambda j :mapp[j][0], x_cord(x))) # blue for true 10s
        x_cord_f = lambda g : list(map(lambda t : t, list(filter(lambda s : mapp[s][1] == g == 1 - tru(s), kk))))# green for false non-10s
        y_cord_f = lambda x : list(map(lambda j :mapp[j][0], x_cord_f(x)))# yellow for false 10s
        plt.figure(figsize = (8,6))
        plt.plot(x_cord(0), y_cord(0) ,'r+', x_cord(1), y_cord(1), 'bo', x_cord_f(0), y_cord_f(0), 'g+', x_cord_f(1), y_cord_f(1), 'yo')
        plt.legend(('TN', 'TP', 'FN', 'FP'), loc = 0)
        plt.xlabel('datapoints')
        plt.ylabel('Activations')
        accuracy([tru(i) for i in kk], [mapp[i][1] for i in kk], len(y_cord(0)), len(y_cord(1)), len(y_cord_f(1)), len(y_cord_f(0)))
    else:# just to see where the actual activations lighten up for the last layer
        x_cord = list(filter(lambda j : j%10==0, kk))# red for 10s
        y_cord = list(map(lambda j :mapp[j], x_cord))      
        x_cord_f = list(filter(lambda j : j%10!=0, kk))
        y_cord_f = list(map(lambda j : mapp[j], x_cord_f))#green for non 10s
        plt.figure(figsize = (8,6))
        plt.plot(x_cord, y_cord, 'r+', x_cord_f, y_cord_f, 'g+')
        plt.legend(('10s', 'non-10s'), loc = 'best') 
        plt.xlabel('Data points')
        plt.ylabel('Activations')

def setup(nn_arch, epochs, gd, thresh):
    weights, bias = model_trainer(gd, epochs, nn_arch) 
    mapp = testing(weights, bias, len(nn_arch), thresh)
    graph(mapp, thresh)
```
